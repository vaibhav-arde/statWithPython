##Stadardization

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
mean,std
```

```python
normalized_data = []
for i in total_bill:
  z_score=(i-mean)/std
  normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.histplot(total_bill)
```

```python
sns.histplot(normalized_data)
```

## Nominal | One Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue', 'green', 'green', 'red', 'blue']
})
```

```python
df
```

```python
# Create an instance of one hot encoder
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoder.fit_transform(df[['color']]).toarray(), columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df, encoded_df], axis=1)
```

# Label And Ordinal Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# create a sample dataframe with a categorical variable
df = pd.DataFrame({
    'color':['red','blue', 'green', 'green', 'red', 'blue']
})
```

```python
df
```

```python
# create an instance of labelencoder
encoder = LabelEncoder()
```

```python
encoder.fit_transform(df['color'])
```

```python
# Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

# create a sample dataframe with a categorical variable
df = pd.DataFrame({
    'size':['small','medium', 'large', 'medium', 'small', 'large']
})
```

```python
list(df['size'].unique())
```

```python
unique_sizes = df['size'].unique()
```

```python
# create an instance of the ordinalEncoder class and fit_transform
encoder = OrdinalEncoder(categories=[unique_sizes])
```

```python
encoder.fit_transform(df[['size']])
```

# Target Guide Original Encoding

```python
import pandas as pd
```

```python
#Create a sample dataframe
df = pd.DataFrame({
      "city": [ "New York", "London", "Paris", "Tokyo", "New York", "Paris"],
          "price": [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
df.groupby('city')
```

```python
#Calcualte the mean price for each city
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
df
```
